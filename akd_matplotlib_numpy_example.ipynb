```python
##Sample graph using matplotlib library and numpy

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=np.linspace(0,5,11)
print (x)
y=x**2
print (y)
#functional method
plt.plot(x,y)
plt.show()
#Another example
plt.plot(x,y)
plt.xlabel('X-Label')
plt.ylabel('Y-label')
plt.title('my Graph')
plt.show()
#Another one
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.subplot(1,2,2)
plt.plot(y,x,'b')

#OO
fig=plt.figure()
axes =fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_title('Perfect')

#Graph within graph
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.20,0.4,0.3,0.4])
axes1.plot(x,y)
axes1.set_xlabel('X-Large')
axes1.set_ylabel('Y-Large')
axes1.set_title('Large')
axes2.plot(y,x)
axes2.set_title('Small')
axes2.set_xlabel('X_small')
axes2.set_ylabel('Y-small')
```
